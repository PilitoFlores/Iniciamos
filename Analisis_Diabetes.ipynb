```python
# An√°lisis de Agrupamientos (Clustering)
# Importar librer√≠as necesarias
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Seleccionar solo las columnas num√©ricas para el clustering
X = df[['weight', 'height', 'blood_glucose', 'physical_activity', 'bmi']]

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar el m√©todo del codo para encontrar el n√∫mero √≥ptimo de clusters
inertia = []
for k in range(1, 11):  # Probar con diferentes valores de k (clusters)
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Graficar el m√©todo del codo
plt.plot(range(1, 11), inertia, marker='o')
plt.title('M√©todo del Codo para K-means')
plt.xlabel('N√∫mero de clusters (K)')
plt.ylabel('Inercia')
plt.show()
```

```python
# Aplicar K-means (3) y visualizar los resultados
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# Asignar los clusters a las observaciones
df['cluster'] = kmeans.labels_

# Ver los primeros registros con los clusters asignados
print(df[['user_id', 'weight', 'height', 'blood_glucose', 'bmi', 'cluster']].head())

# Graficar los clusters en funci√≥n de dos caracter√≠sticas (peso y glucosa en sangre)
plt.figure(figsize=(8, 6))
plt.scatter(df['weight'], df['blood_glucose'], c=df['cluster'], cmap='viridis')
plt.title('Clustering de los Datos de Diabetes')
plt.xlabel('Peso (kg)')
plt.ylabel('Glucosa en Sangre (mg/dL)')
plt.colorbar(label='Cluster')
plt.show()
```
