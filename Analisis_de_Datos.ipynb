```python
# Identificación de Outlier
# Función para identificar valores atípicos utilizando el rango intercuartílico (IQR)
def detectar_outliers_iqr(df,columnas):
    resumen_outliers={}
    for col in columnas:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1
        limite_inferior=Q1-1.5*IQR
        limite_superior=Q3+1.5*IQR
        outliers=df[(df[col]<limite_inferior)|(df[col]>limite_superior)]
        resumen_outliers[col]={
            'IQR':IQR,
            'Límite Inferior':limite_inferior,
            'Límite Superior':limite_superior,
            'Total Outliers':len(outliers),
        }
    return pd.DataFrame(resumen_outliers).T
```

```python
# Identificar Valores Atípicos en las Variables Numéricas Clave
variables_clave=['SALES','QUANTITYORDERED','PRICEEACH']
resumen_outliers=detectar_outliers_iqr(df,variables_clave)
print("Resumen de Outliers:\n")
print(resumen_outliers)
```

```python
# Eliminar outliers de SALES y QUANTITYORDERED
# Filtrar datos eliminando outliers basados en el IQR
def eliminar_outliers(df, columnas):
    for col in columnas:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        df = df[(df[col] >= limite_inferior) & (df[col] <= limite_superior)]
    return df

# Crear un nuevo DataFrame sin outliers
df_sin_outliers = eliminar_outliers(df, ['SALES', 'QUANTITYORDERED'])

print(f"Tamaño del dataset original: {df.shape}")
print(f"Tamaño del dataset sin outliers: {df_sin_outliers.shape}")
```

```python
# Ajustar los valor de SALES y QUANTITYORDERED
def winsorizar(df, columnas):
    for col in columnas:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower=limite_inferior, upper=limite_superior)
    return df

# Aplicar winsorización
df_winsorizado = winsorizar(df.copy(), ['SALES', 'QUANTITYORDERED'])

# Comparar con el original
print("Dataset después del ajuste:")
print(df_winsorizado[['SALES', 'QUANTITYORDERED']].describe())
```

```python
# Análisis Específico de Outliers
# Extraer outliers para análisis separado
def extraer_outliers(df, columnas):
    outliers = pd.DataFrame()
    for col in columnas:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        outliers_col = df[(df[col] < limite_inferior) | (df[col] > limite_superior)]
        outliers = pd.concat([outliers, outliers_col])
    return outliers.drop_duplicates()

# Crear un DataFrame con los outliers
outliers = extraer_outliers(df, ['SALES', 'QUANTITYORDERED'])
print(f"Cantidad de outliers detectados: {outliers.shape[0]}")
```

```python
# Clasificar Clientes en Base a la Probabilidad de Recompra
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Crear etiquetas para recompra (1: Sí, 0: No) basado en la última fecha
df['Recompra'] = (df['ORDERDATE'] > '2004-12-31').astype(int)

# Variables predictoras
X = df[['SALES', 'QUANTITYORDERED', 'PRICEEACH']] 
y = df['Recompra']

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar un modelo Random Forest
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predicciones
y_pred = clf.predict(X_test)

# Evaluación del modelo
print("Reporte de Clasificación:\n", classification_report(y_test, y_pred))

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Visualización de la matriz de confusión con tamaño personalizado
fig, ax = plt.subplots(figsize=(3, 3))  
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot(cmap='Blues_r', colorbar=True, ax=ax)  
plt.title("Matriz de Confusión")
plt.show()
```

## **Clasificación de Clientes según Recompra (Random Forest)**

- *Propósito*: Identificar clientes con alta probabilidad de realizar una recompra usando características clave como ventas, cantidad ordenada y precio unitario.

- *Observaciones de la Matriz de Confusión*:

          - Clase 0 (No Recompra):

               * Alta precisión (84%) y buen recall (93%), indicando que el modelo identifica correctamente a la mayoría de los clientes que no recompran.

          - Clase 1 (Recompra):

               * Precisión baja (38%) y recall (19%), lo que indica que el modelo tiene dificultades para identificar clientes que recompran.

*Conclusión*:

- El modelo es más efectivo para predecir clientes que no recompran que aquellos que sí lo hacen.

```python
# Segmentación Adicional para identificar patrones ocultos en las ventas y el comportamiento de los clientes
# Calcular inercia para diferentes números de clústeres - Método del Codo
inertias = []
for k in range(1, 11):  # Prueba entre 1 y 10 clústeres
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)  # Usa los datos escalados
    inertias.append(kmeans.inertia_)

# Graficar el Método del Codo
plt.figure(figsize=(6, 2))
plt.plot(range(1, 11), inertias, marker='o', linestyle='--')
plt.title('Método del Codo')
plt.xlabel('Número de Clústeres')
plt.ylabel('Inercia')
plt.grid()
plt.show()
```

Método del Codo:

La gráfica del método del codo muestra que la inercia disminuye significativamente entre 2 y 4 clústeres, pero la tasa de reducción comienza a estabilizarse a partir de 4. Esto sugiere que 4 clústeres podría ser un buen punto de equilibrio.

```python
# Calcular el Silhouette Score para diferentes números de clústeres - Coeficiente de Silueta
silhouette_scores = []
for k in range(2, 11):  # El Silhouette no se calcula para k=1
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    score = silhouette_score(df_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Graficar el Coeficiente de Silueta
plt.figure(figsize=(6, 2))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('Coeficiente de Silueta')
plt.xlabel('Número de Clústeres')
plt.ylabel('Puntaje de Silueta')
plt.grid()
plt.show()
```

Coeficiente de Silueta:

El coeficiente de silueta es más alto para 2 clústeres (~0.36). Sin embargo, elegir 2 clústeres podría no capturar suficiente variabilidad de los datos. Con 4 clústeres, el puntaje de silueta no es el mejor, pero aún es razonable (~0.33), lo cual respalda la decisión del método del codo.
